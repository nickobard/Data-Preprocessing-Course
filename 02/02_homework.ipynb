```python
import pandas as pd
import numpy as np
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from matplotlib.colors import ListedColormap
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification, make_moons, make_circles
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split, ParameterGrid
import sklearn.metrics as metrics
from tqdm import tqdm

dataset_path = 'data.csv'
rd_seed = 42
```

The dataset can be considered as big. Thus, when splitting data into train-valid-test sets, we don't need large validation and test sets. We will use 80-10-10 ratio. 

```python
def split_data(data):
    print(f'Data shape: {data.shape}')
    labels = data['y']
    features = data.drop('y', axis=1)
    X_train, X_rest, y_train, y_rest = train_test_split(features, labels, train_size=0.8, random_state=rd_seed)
    X_val, X_test, y_val, y_test = train_test_split(X_rest, y_rest, test_size=0.5, random_state=rd_seed)
    print(f'Train shape: {X_train.shape}')
    print(f'Validation shape: {X_val.shape}')
    print(f'Test shape: {X_test.shape}')
    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = split_data(df)
```

## 2. Classification Algorithms


For classification task we will use Support Vector Machine - Linear and RBF.

We will use some visualization of decision border on synthetic data using function from here: https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html

This will give us rough understanding of how chosen models separate data.

```python
def plot_classifiers_borders():
    names = [
        "Linear SVM",
        "RBF SVM"
    ]

    classifiers = [
        LinearSVC(C=0.025, random_state=rd_seed),
        SVC(gamma=2, C=1, random_state=rd_seed)
    ]

    X, y = make_classification(
        n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
    )
    # noise generator
    rng = np.random.RandomState(rd_seed)
    # add some noise along X axis
    X += 2 * rng.uniform(size=X.shape)
    linearly_separable = (X, y)

    datasets = [
        make_moons(noise=0.3, random_state=0),
        make_circles(noise=0.2, factor=0.5, random_state=1),
        linearly_separable
    ]

    figure = plt.figure(figsize=(9, 9), constrained_layout=True)
    i = 1
    # iterate over datasets
    for ds_cnt, ds in enumerate(datasets):
        # preprocess dataset, split into training and test part
        X, y = ds
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.4, random_state=rd_seed
        )

        x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
        y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

        # just plot the dataset first
        cm = plt.cm.RdBu
        cm_bright = ListedColormap(["#FF0000", "#0000FF"])
        ax = figure.add_subplot(len(datasets), len(classifiers) + 1, i)
        if ds_cnt == 0:
            ax.set_title("Input data")
        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
        # Plot the testing points
        ax.scatter(
            X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.4, edgecolors="k"
        )
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xticks(())
        ax.set_yticks(())
        i += 1

        # iterate over classifiers
        for name, clf in zip(names, classifiers):
            ax = figure.add_subplot(len(datasets), len(classifiers) + 1, i)

            clf = make_pipeline(StandardScaler(), clf)
            clf.fit(X_train, y_train)
            score = clf.score(X_test, y_test)
            DecisionBoundaryDisplay.from_estimator(
                clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
            )

            # Plot the training points
            ax.scatter(
                X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
            )
            # Plot the testing points
            ax.scatter(
                X_test[:, 0],
                X_test[:, 1],
                c=y_test,
                cmap=cm_bright,
                edgecolors="k",
                alpha=0.6,
            )

            ax.set_xlim(x_min, x_max)
            ax.set_ylim(y_min, y_max)
            ax.set_xticks(())
            ax.set_yticks(())
            if ds_cnt == 0:
                ax.set_title(name)
            ax.text(
                x_max - 0.3,
                y_min + 0.3,
                ("%.2f" % score).lstrip("0"),
                size=15,
                horizontalalignment="right",
            )
            i += 1

    plt.show()


plot_classifiers_borders()
```

Now let's train and finetune linear and RBF models on dataset without preprocessing

### SVM with Linear Kernel

```python
C = np.append(np.linspace(1e-2, 1, 5), np.linspace(1, 10, 6)[1:])
param_grid_dict = {
    'C': C
}

param_grid = ParameterGrid(param_grid_dict)
```

```python
def train(model_class):
    acc_train = []
    acc_val = []
    for params in tqdm(param_grid):
        model = model_class(**params, random_state=rd_seed)
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_val_pred = model.predict(X_val)
        acc_train.append(metrics.accuracy_score(y_train, y_train_pred))
        acc_val.append(metrics.accuracy_score(y_val, y_val_pred))
    return acc_train, acc_val
```

```python
acc_train, acc_val = train(LinearSVC)
```

```python
def plot_accuracy():
    fig, ax = plt.subplots(figsize=(20, 7))
    ax.plot(acc_train, 'o-', label='train', color='red')
    ax.plot(acc_val, 'o-', label='validation', color='navy')
    ax.set_xlabel('C value')
    ax.set_ylabel('Accuracy')
    ax.set_title('Model Accuracy by C hyperparameter')
    ax.legend(loc="lower right")
```

```python
plot_accuracy()
```

### SVM with RBF Kernel

```python
param_grid_dict['kernel'] = ['rbf']
param_grid = ParameterGrid(param_grid_dict)
acc_train, acc_val = train(SVC)
plot_accuracy()
```
